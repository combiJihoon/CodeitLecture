```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

records = []
num = 0
response = requests.get(
    'https://workey.codeit.kr/ratings/index?year=2010&month=1&weekIndex='+str(num))
soup = BeautifulSoup(response.text, 'html.parser')

period_tag = soup.select(
    'div.search-box > section.header-row > select#weekSelectBox > option')
rank_tag = soup.select('table > tr > td.rank')
channel_tag = soup.select('table > tr > td.channel')
program_tag = soup.select('table > tr > td.program')
rating_tag = soup.select('table > tr > td.percent')

for i in range(len(rank_tag)):
    record = []
    record.append(period_tag[0].text)
    record.append(rank_tag[i].text)
    record.append(channel_tag[i].text)
    record.append(program_tag[i].text)
    record.append(rating_tag[i].text)
    records.append(record)


print(records)

df = pd.DataFrame(data=records, columns=[
                  'period', 'rank', 'channel', 'program', 'rating'])
```
