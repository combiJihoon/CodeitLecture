```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('Data/iphone.csv', index_col = 0)
df
```

```python
df['디스플레이'] > 5
```

```python
df.loc[df['디스플레이'] > 5] 
```

```python
df.loc[df['디스플레이'] > 5] = 'p'
df
```

디스플레이가 5보다 크면 모두 p로 변경함

```python
df.iloc[[1,3], [1,4]] = 'v'
df
```

```python
df.loc['iPhone XR'] = ['1', '2', '3', '4', '5']
df
```

iPhone XR이 추가됨

```python
df['제조사'] = 'Apple'
df
```

열을 추가하는 방법(without loc)

```python
df.drop('iPhone XR', axis = 'index', inplace = False)
```

drop에 axis = 'index'를 쓰면 컬럼이 아니라 로우를 삭제하고 싶다는 뜻(판다스에서 로우는 인덱스이기도 하니까)

inplace = False를 쓰면 기존 데이터인 df 자체는 건드리지 않겠다는 뜻 -> 따라서 iPhone XR을 삭제한 데이터프레임은 새로운 변수에 저장할 것

```python
df
```

```python
df.drop('제조사', axis = 'columns', inplace = True)
df
```
