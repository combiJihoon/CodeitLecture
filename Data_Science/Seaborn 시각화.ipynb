```python
import pandas as pd
import seaborn as sns
```

```python
body_df = pd.read_csv('Data/body.csv', index_col=0)
```

```python
body_df.head()
```

```python
body_df['Height']
```

```python
body_df['Height'].value_counts()
```

키는 Height 시리즈의 인덱스에 해당하므로 인덱스 순으로 정렬하라는 명령을 주면 됨

```python
body_df['Height'].value_counts().sort_index()
```

```python
body_df['Height'].value_counts().sort_index().plot()
```

seaborn library에서 kdeplot이라는 함수를 불러오고 파라미터에 body_df['Height']를 넣어주면 된다.

```python
sns.kdeplot(body_df['Height'])
```

```python
sns.kdeplot(body_df['Height'], bw = 0.5)
```

```python
sns.kdeplot(body_df['Height'], bw = 0.05)
```

bw는 bin width의 약자인듯 함 -> 곡선이 부드러운 정도를 나타냄

```python
body_df.plot(kind ='hist', y = 'Height', bins = 15)
```

```python
sns.distplot(body_df['Height'])
```

 히스토그램 위에 kdeplot이그려짐

```python
body_df.plot(kind='box', y = 'Height')
```

```python
sns.violinplot(y = body_df['Height'])
```

violinplot은 분포 전체를 보여준다는 장점이 있음

```python
body_df.plot(kind='scatter', x = 'Height', y = 'Weight')
```

```python
sns.kdeplot(body_df['Height'], body_df['Weight'])
```

kdeplot( , ) 앞쪽은 x축, 뒤쪽은 y축을 나타냄

등고선처럼 생각하면 된다. 선끼리 가까우면 값이 많다는 뜻이며 키와 몸무게에 대한 kdeplot을 2차원으로 나타낸 것임

```python
sns.lmplot(data=body_df, x = 'Height', y = 'Weight')
```

회귀선 = 흩어진 점들을 하나의 선으로 표현한 것

lmplot의 lm은 linear model의 약자인듯함~

```python
laptops_df = pd.read_csv('Data/laptops.csv')
laptops_df
```

```python
laptops_df['os']
```

```python
laptops_df['os'].unique()
```

```python
sns.catplot(data=laptops_df, x = 'os', y = 'price')
```

 카테고리별로 비교하려면 catplot을 활용하면 됨 -> 예를 들어, 어떤 os가 가장 비싼지 확인하고 싶을 때

```python
sns.catplot(data=laptops_df, x = 'os', y = 'price', kind = 'box')
```

```python
sns.catplot(data=laptops_df, x = 'os', y = 'price', kind = 'violin')
```

```python
sns.catplot(data=laptops_df, x = 'os', y = 'price', kind = 'strip')
```

데이터 개수가 몇개인지 모르므로 위의 박스 플롯이 얼마나 믿을만한지 알 수 없다 -> 따라서 이럴 땐 strip을 쓰면 된다(기본 형태인듯)

```python
laptops_df['processor_brand'].unique()
```

위 그래프에서 어떤 것들이 인텔을 쓰고 어떤 것들이 amd를 쓰는지 파악할 수 있음

```python
sns.catplot(data=laptops_df, x = 'os', y = 'price', kind = 'strip',  hue = 'processor_brand')
```

hue는 색을 의미함 -> 즉, processor_brand에 따라 색을 달리하는 것

```python
sns.catplot(data=laptops_df, x = 'os', y = 'price', kind = 'swarm',  hue = 'processor_brand')
```

점들이 너무 몰려있어 보기 불편할 때 swarm plot을 사용함

```python
exam_df = pd.read_csv('Data/exam.csv')
exam_df
```

```python
exam_df.corr()
```

corr()은 상관계수를 나타내며 데이터프레임으로 나옴 -> 따라서 이를 보기 좋게 seaborn 그래프로 그리면 좋음

```python
sns.heatmap(exam_df.corr())
```

색이 밝을수록 상관관계가 높다는 뜻임 -> reading score와 writing score의 상관계수가 높음

```python
sns.heatmap(exam_df.corr(), annot = True)
```

annot = True를 쓰면 상관계수 숫자도 포함시켜줌
